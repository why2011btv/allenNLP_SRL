```python

# Python3 program to find the  
# longest path in the DAG  
    
# Function to traverse the DAG  
# and apply Dynamic Programming  
# to find the longest path  
finishing_time = []
def dfs(node, adj, dp, vis):  
    print("Running dfs for node", node)
    # Mark as visited  
    vis[node] = True 
    
    # Traverse for all its children  
    for i in range(0, len(adj[node])):   
        print("Now handling node", node, "'s neighbor:", adj[node][i])
        #print("Now in for loop of node", node)
        # If not visited  
        if not vis[adj[node][i]]: 
            print("node", node, "'s neighbor", adj[node][i], "is not visited, run dfs for it")
            dfs(adj[node][i], adj, dp, vis)  
        else:
            print("node", node, "'s neighbor", adj[node][i], "has been visited")
        # Store the max of the paths  
        
        dp[node] = max(dp[node], 1 + dp[adj[node][i]]) 
        print(dp)
    print("FOR loop of node", node, "finished")
    finishing_time.append(node)
# Function to add an edge  
def addEdge(adj, u, v):  
   
    adj[u].append(v)  
    
# Function that returns the longest path  
def findLongestPath(adj, n):  
   
    # Dp array  
    dp = [0] * (n + 1)  
      
    # Visited array to know if the node  
    # has been visited previously or not  
    vis = [False] * (n + 1) 
      
    # Call DFS for every unvisited vertex  
    for i in range(0, n + 1):   
        if not vis[i]:  
            dfs(i, adj, dp, vis)  
       
    ans = 0 
    
    # Traverse and find the maximum of all dp[i]  
    for i in range(0, n + 1):   
        ans = max(ans, dp[i])  
    print(dp)   
    return ans  
   
# Driver Code  
#if __name__ == "__main__":  
if 1:   
    n = 8
    adj = [[] for i in range(n + 1)] 
    
    # Example-1  
    addEdge(adj, 0, 1)  
    addEdge(adj, 1, 2)  
    addEdge(adj, 2, 3)  
    addEdge(adj, 3, 0)  
    addEdge(adj, 2, 4)
    addEdge(adj, 4, 5)
    addEdge(adj, 5, 6)
    addEdge(adj, 6, 4)
    addEdge(adj, 7, 6)
    addEdge(adj, 7, 8)
    
    
    print(findLongestPath(adj, n)) 
    print(finishing_time)
    
# This code is contributed by Rituraj Jain 
```
